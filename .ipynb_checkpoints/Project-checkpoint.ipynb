```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Importing data
df = pd.read_csv('HOtel Reservations.csv')
df.columns
```

```python
#missing data
for colName, colData in df.items():
    print(colName, ':', df[colName].isna().sum())
```

```python
#question, predict whether or not customer will cancel his reservation
#the target variable
print(df.booking_status.value_counts(dropna=False))
#replace variable
df.booking_status.replace('Not_Canceled',0, inplace=True)
df.booking_status.replace('Canceled',1, inplace=True)
#df.booking_status.replace(0, 'Not_Canceled', inplace=True)
#df.booking_status.replace(1, 'Canceled', inplace=True)
df.dtypes
```

```python
#varibles prone to one-hot encoding
print(df.type_of_meal_plan.value_counts())
print(df.room_type_reserved.value_counts())
print(df.market_segment_type.value_counts())
```

```python
#visualizing correlations
correlations = df.corr(numeric_only=True).booking_status
correlations = correlations.abs().sort_values(ascending=False)
print(correlations)
ax = correlations.plot(kind='bar')
ax.set(ylim=[0, 1], ylabel='pearson correlation')
```

```python
#we can see that the first 5 variables are the most significant for analysis
features = correlations.index[:6]
df_features = df[features]
df_features['arrival_month'] = df['arrival_month']
df_features
```

```python
#start building a model
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
df_features.dtypes
```

```python
#predictors and target variable
#y = df.booking_status
#X = df.drop(['booking_status', 'Booking_ID'], axis=1)

#preprocessing data

#One hot encoding
#X = pd.get_dummies(X, prefix=['meal', 'room_type', 'segment'],columns=['type_of_meal_plan', 'room_type_reserved', 'market_segment_type'])

#train-test split
#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

#standardscaler
```

```python
#predictors and target variable
#removing arrival_year, using instead month for season estimator.
y = df_features.booking_status
X = df_features.drop(['booking_status','arrival_year'], axis=1)

#One hot encoding
X = pd.get_dummies(X, prefix=['arrival_month'] ,columns=['arrival_month'])

#preprocessing data
#train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

#standardscaler
scaler = StandardScaler().fit(X_train)
X_train_scl = scaler.transform(X_train)
X_test_scl = scaler.transform(X_test)
```

```python
#lists to stores the scores
training_scores = []
test_scores = []
```

```python
#Building the first model Logistic regression
from sklearn.linear_model import LogisticRegressionCV

lr_l2 = LogisticRegressionCV(Cs=10, cv=5, penalty='l2', solver='sag')
lr_l2.fit(X_train, y_train)

y_pred_train = lr_l2.predict(X_train)
y_pred_test = lr_l2.predict(X_test)
print('train_score l2 : ', lr_l2.score(X_train, y_train))
print('test_score l2 : ', lr_l2.score(X_test, y_test))

#from sklearn.metrics import accuracy_score
#print('Also test_score', accuracy_score(y_test, y_pred_test))
```

```python
coefficients = lr_l2.coef_
features_names = X_train.columns

# get the corresponding feature names
feature_coefficients = list(zip(features_names, coefficients[0]))

# print the feature names and their respective coefficients
for feature, coefficient in feature_coefficients:
    print(f'{feature}: {coefficient}')
```

```python
from sklearn.model_selection import GridSearchCV
parameters = {'penalty' : ['l1', 'l2', 'elasticnet'], 
              'solver': ['lbfgs', 'liblinear', 'sag', 'saga']}
lr = LogisticRegressionCV()
grid = GridSearchCV(estimator=lr, param_grid = parameters)
grid.fit(X_train, y_train)
```

```python
#a function to get grid search results
def best_param(obj):
    results = pd.DataFrame(obj.cv_results_)
    indx = results['mean_test_score'].idxmax()
    #print('indx max score : ', indx)
    return pd.DataFrame(results.iloc[indx,:]).T
```

```python
#best parameters for estimator
print('Best mean_test_score (CV) : ', grid.best_score_) # this for CV in grid algorithm
best_param(grid)
```

```python
#The best parameters seem to be the previously choosen
#A simpler way to get the parameters
print('Best parameters : ', grid.best_params_)
print('train_score : ', grid.score(X_train, y_train))
print('test_score : ', grid.score(X_test, y_test))

training_scores.append(grid.score(X_train, y_train))
test_scores.append(grid.score(X_test, y_test))
```

```python
#Trying a second algorith KNN
from sklearn.neighbors import KNeighborsClassifier

kneigh = KNeighborsClassifier(n_neighbors=5)
kneigh.fit(X_train, y_train)
print('train_score : ', kneigh.score(X_train, y_train))
print('test_score : ', kneigh.score(X_test, y_test))
```

```python
#Grid search for different parameters
params = {'n_neighbors' : [3,4,5,6,]}
kneigh = KNeighborsClassifier()
grid = GridSearchCV(estimator=kneigh, param_grid=params)
grid.fit(X_train, y_train)
```

```python
#Scores:
print('Best parameters : ', grid.best_params_)
print('train_score : ', grid.score(X_train, y_train))
print('test_score : ', grid.score(X_test, y_test))

training_scores.append(grid.score(X_train, y_train))
test_scores.append(grid.score(X_test, y_test))
```

```python
#Training the third algorithm: Extra trees classifier
from sklearn.ensemble import ExtraTreesClassifier

clf = ExtraTreesClassifier(n_estimators=100, max_depth=2)
clf.fit(X_train, y_train)
print('train_score : ', clf.score(X_train, y_train))
print('test_score : ', clf.score(X_test, y_test))

training_scores.append(grid.score(X_train, y_train))
test_scores.append(grid.score(X_test, y_test))
```

```python
#Using grid search
from sklearn.model_selection import GridSearchCV
clf = ExtraTreesClassifier()
params = {'n_estimators' : [25,50,75,100,125], 'max_depth' : [3,4,None]}
grid = GridSearchCV(estimator = clf, param_grid = params)
grid.fit(X_train, y_train)
```

```python
#Scores:
print('Best parameters : ', grid.best_params_)
print('train_score : ', grid.score(X_train, y_train))
print('test_score : ', grid.score(X_test, y_test))

training_scores.append(grid.score(X_train, y_train))
test_scores.append(grid.score(X_test, y_test))
```

```python
# It is worth it to try Gradient boosting
from sklearn.ensemble import GradientBoostingClassifier

clf = GradientBoostingClassifier(learning_rate=0.1, n_estimators=100)
clf.fit(X_train, y_train)

print('train_score : ', clf.score(X_train, y_train))
print('test_score : ', clf.score(X_test, y_test))
```

```python
#trying gridsearch
clf = GradientBoostingClassifier()
param = {'learning_rate' : [0.1, 1], 'n_estimators' : [500, 600, 700]}
grid = GridSearchCV(estimator=clf, param_grid=param)
grid.fit(X_train, y_train)
```

```python
print('Best parameters : ', grid.best_params_)
print('train_score : ', grid.score(X_train, y_train))
print('test_score : ', grid.score(X_test, y_test))

training_scores.append(grid.score(X_train, y_train))
test_scores.append(grid.score(X_test, y_test))
```

```python
#gradient boosing seems to be the best performing model
from sklearn.ensemble import GradientBoostingClassifier

clf = GradientBoostingClassifier(learning_rate=1, n_estimators=600)
clf.fit(X_train, y_train)

print('train_score : ', clf.score(X_train, y_train))
print('test_score : ', clf.score(X_test, y_test))
```

```python
y_test_predict = clf.predict(X_test)
df_results = X_test.copy()
df_results['y_test'] = y_test
df_results['y_test_predict'] = y_test_predict
#print(df_results)

from sklearn.metrics import classification_report
print(classification_report(y_test, y_test_predict))

from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_test_predict))
```

```python
#The results:
data = {'Algorithm': ['Logistic_Regression', 'KNN', 'ExtraTreesClassifier', 'Gradient_Boosting'],
        'Train_score': [0.7785129174543163, 0.87338531821046, 0.9862161310649024, 0.9296235034656585],
        'Test_score': [0.7879261233115868, 0.8236699439492787, 0.86244601672333, 0.8622622438665809]}
df_scores = pd.DataFrame(data)
df_scores
```
